```python
import pandas as pd

from pyod.models.ecod import ECOD
```

```python
df = pd.read_csv("data/train.csv", sep = ";")
df = df.iloc[:, 0:8]

df
```

```python
data = pd.read_parquet('data/df_preprocessed.parquet')

data
```

```python
# https://github.com/yzhao062/pyod

clf = ECOD()

clf.fit(data)
outliers = clf.predict(data) 

data["outliers"] = outliers
df["outliers"] = outliers
```

```python
print(f"Without Outliers : {data[data.outliers == 0].shape[0]:,}")
print(f"Original         : {data.shape[0]:,}")
```

```python
# Data without outliers
data_no_outliers = data[data["outliers"] == 0]
data_no_outliers = data_no_outliers.drop(["outliers"], axis=1)

data_no_outliers
```

```python
# Original Data without Outliers
df_no_outliers = df[df["outliers"] == 0]
df_no_outliers = df_no_outliers.drop(["outliers"], axis=1)

df_no_outliers
```

```python
data_no_outliers.to_parquet('data/data_no_outliers.parquet', index=False)
df_no_outliers.to_parquet('data/df_no_outliers.parquet', index=False)
```
